```python
import os
import pandas as pd
import json

# Define the base directory
base_dir = r"C:\Users\sharm\OneDrive\Desktop\Kishan\Contractzy\WebScrapping\Tutorial\Financial_Data\MoneyControl\Companies\IT Services & Consulting"
# Output JSON file
output_file = "ITSector.json"
```

```python
# Initialize dictionary to store data
it_sector_data = {}

# Iterate through each company's folder
for company in os.listdir(base_dir):
    company_path = os.path.join(base_dir, company)
    excel_path = os.path.join(company_path, "Pruned_Excel", "Final_Parameters", f"{company}_semi_final.xlsx")
    
    if os.path.isfile(excel_path):
        # Load Excel file
        xls = pd.ExcelFile(excel_path)
        
        # Dictionary to hold all sheet data for the company
        company_data = {}
        
        # Read all five required sheets
        for sheet in ['Balance_Sheet', 'Cash_Flow', 'Profit_Loss', 'Quarterly', 'Ratio']:
            if sheet in xls.sheet_names:
                df = xls.parse(sheet)
                
                # Extract years from the first row (assuming they are column headers after the first one)
                years = df.columns[1:].tolist()
                
                # Transform data into structured format
                sheet_data = {"Years": years}
                for index, row in df.iterrows():
                    key = row.iloc[0]  # First column contains metric names
                    values = row.iloc[1:].tolist()  # The rest are yearly values
                    sheet_data[key] = values
                
                company_data[sheet] = sheet_data
        
        # Add company data to the main dictionary
        it_sector_data[company] = company_data

# Save to JSON file
with open(output_file, "w", encoding="utf-8") as json_file:
    json.dump(it_sector_data, json_file, indent=4)

print(f"JSON file '{output_file}' has been created successfully.")
```

```python
import os
import json
import pandas as pd

# Base directory containing all company folders
base_dir = r"C:\Users\sharm\OneDrive\Desktop\Kishan\Contractzy\WebScrapping\Tutorial\Financial_Data\MoneyControl\Companies\IT Services & Consulting"

# Output JSON file
output_json_path = "ITSector2.json"

# Dictionary to store all company data
all_companies_data = {"Companies": {}}

# Iterate through each company folder
for company in os.listdir(base_dir):
    company_path = os.path.join(base_dir, company)
    excel_path = os.path.join(company_path, "Pruned_Excel", "Final_Parameters", f"{company}_semi_final.xlsx")
    
    # Check if the Excel file exists
    if os.path.exists(excel_path):
        print(f"Processing: {company}")
        
        # Load the Excel file
        xls = pd.ExcelFile(excel_path)
        
        # Dictionary to store this company's data
        company_data = {}
        
        # Extract data from each sheet
        for sheet_name in xls.sheet_names:
            df = pd.read_excel(xls, sheet_name=sheet_name)
            company_data[sheet_name] = df.to_dict(orient="list")
        
        # Add company's data to the main dictionary
        all_companies_data["Companies"][company] = company_data
    else:
        print(f"Skipping {company} (Excel file not found)")

# Save the combined JSON output
with open(output_json_path, "w") as json_file:
    json.dump(all_companies_data, json_file, indent=4)

print(f"All company data saved to {output_json_path}")
```

```python
import pandas as pd
import json

def excel_to_json(excel_path, json_output_path):
    # Load the Excel file
    xls = pd.ExcelFile(excel_path)
    
    # Extract data from each sheet and store it in column-wise JSON format
    data_dict = {}
    for sheet_name in xls.sheet_names:
        df = pd.read_excel(xls, sheet_name=sheet_name)
        data_dict[sheet_name] = df.to_dict(orient="list")
    
    # Save the JSON output
    with open(json_output_path, "w") as json_file:
        json.dump(data_dict, json_file, indent=4)
    
    print(f"JSON data saved to {json_output_path}")

# Example usage
excel_path = r"C:\Users\sharm\OneDrive\Desktop\Kishan\Contractzy\WebScrapping\Tutorial\Financial_Data\MoneyControl\Companies\IT Services & Consulting\3i Infotech Ltd\Pruned_Excel\Final_Parameters\3i Infotech Ltd_semi_final.xlsx"  # Update with your file path
json_output_path = "3i_Infotech_data.json"
excel_to_json(excel_path, json_output_path)
```
