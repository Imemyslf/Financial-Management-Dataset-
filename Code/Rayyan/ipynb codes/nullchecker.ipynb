```python
import os
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font

# Define the base directory containing all company folders
input_file = r"C:\Users\rayya\OneDrive\Desktop\Programs\Financial-Management-Dataset-\Financial_Data\MoneyControl\Companies\IT Services & Consulting"
companies_list = os.listdir(input_file)

# Define output file
output_excel = os.path.join(input_file, "Zero_Values_Report.xlsx")

# Create a new Excel writer object
with pd.ExcelWriter(output_excel, engine='openpyxl') as writer:

    # Iterate over each company folder
    for company in companies_list:
        file_path = os.path.join(input_file, company, "Pruned_Excel", "Final_Parameters", "Semi_Final")

        # Define file name
        excel_file = os.path.join(file_path, f"{company}_Semi_Final.xlsx")

        # Check if the file exists before processing
        if os.path.exists(excel_file):
            data = []  # List to store zero value data

            # Load all sheets from the Excel file
            dfs = pd.read_excel(excel_file, sheet_name=None)

            # Iterate through each sheet in the Excel file
            for sheet_name, df in dfs.items():
                # Convert non-numeric columns to NaN (avoid counting 0s in text)
                df_numeric = df.apply(pd.to_numeric, errors='coerce')

                # Count occurrences of zero values per column
                zero_counts = (df_numeric == 0).sum()
                zero_columns = zero_counts[zero_counts > 0]  # Only columns with 0s

                # Collect data for zero values
                if not zero_columns.empty:
                    for column, count in zero_columns.items():
                        total_length = df_numeric[column].count() + df_numeric[column].isnull().sum()
                        data.append([sheet_name, column, count, total_length])

            # Create a DataFrame from collected data
            if data:
                df_result = pd.DataFrame(data, columns=["Sheet Name", "Column Name", "Zero Values", "Column Length"])

                # Save to the company sheet in the output file
                df_result.to_excel(writer, sheet_name=company, index=False)

    # Save the file
    writer.close()

# Apply formatting to the "Sheet Name" column
wb = load_workbook(output_excel)

for sheet in wb.sheetnames:
    ws = wb[sheet]

    # Apply bold formatting to "Sheet Name" column
    for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=1):
        for cell in row:
            cell.font = Font(bold=True)

# Save the formatted file
wb.save(output_excel)

print(f"\n Zero values report saved at: {output_excel}")
print("\n Formatting applied: Sheet names in bold")
```

```python
import os
import pandas as pd
import numpy as np

# Define the base directory containing all company folders
input_file = r"C:\Users\sharm\OneDrive\Desktop\Kishan\Contractzy\WebScrapping\Tutorial\Financial_Data\MoneyControl\Companies\IT Services & Consulting"
companies_list = os.listdir(input_file)

# Iterate over each company folder
for company in companies_list:
    file_path = os.path.join(input_file, company, "Pruned_Excel", "Final_Parameters", "Semi_Final")

    # Define file names
    excel_file = os.path.join(file_path, f"{company}_Semi_Final.xlsx")
    output_file = os.path.join(file_path, f"{company}_Semi_Semi_Final.xlsx")  # Final merged output file

    # Check if the file exists before processing
    if os.path.exists(excel_file):
        print(f"\nProcessing Company: {company}")

        # Load all sheets from the Excel file
        dfs = pd.read_excel(excel_file, sheet_name=None)

        # Iterate through each sheet in the Excel file
        for sheet_name, df in dfs.items():
            if sheet_name == "Quarterly":
                print(f"  -> Processing 'Quarterly' sheet")

                # Ensure first column (Date/Index) is string and extract Quarter info
                df.iloc[:, 0] = df.iloc[:, 0].astype(str)
                df['Quarter'] = df.iloc[:, 0].str.extract(r'(Q[1-4])')

                for column in df.columns[1:-1]:  # Skip first (date) and last (quarter helper) columns
                    zero_mask = df[column] == 0  # Find zero values

                    if zero_mask.sum() > 0:  # If any zeros exist
                        print(f"    - Column: {column}, Zero Count: {zero_mask.sum()}")

                        for quarter in df['Quarter'].dropna().unique():
                            quarter_mask = df['Quarter'] == quarter

                            # Calculate mean excluding zeros
                            valid_values = df.loc[quarter_mask & (df[column] != 0), column]
                            if not valid_values.empty:
                                quarter_mean = valid_values.mean()

                                # Replace zero values with rounded mean (3 decimal places)
                                df.loc[quarter_mask & zero_mask, column] = round(quarter_mean, 3)
                                print(f"      - Filled {zero_mask.sum()} zeros in {column} for {quarter} with mean: {quarter_mean:.3f}")

                # Drop the Quarter helper column before saving
                df.drop(columns=['Quarter'], inplace=True, errors='ignore')

            else:
                print(f"\n  Processing Sheet: {sheet_name}")

                # Convert non-numeric columns to NaN (avoid counting 0s in text columns)
                df_numeric = df.apply(pd.to_numeric, errors='coerce')

                for column in df_numeric.columns[1:]:  # Skip the first column (likely date or index)
                    zero_count = (df_numeric[column] == 0).sum()
                    total_length = df_numeric[column].count() + df_numeric[column].isnull().sum()

                    if zero_count > 0:
                        print(f"    - Column: {column}, Zero Count: {zero_count}, Total Length: {total_length}")

                        if zero_count < (total_length / 2):  # Fill only if zeros are <50% of total rows
                            column_mean = df_numeric[column][df_numeric[column] != 0].mean()

                            if not pd.isna(column_mean):  # Avoid replacing with NaN
                                df[column] = df[column].replace(0, round(column_mean, 3))
                                print(f"      -> Filled {zero_count} zeros with mean: {column_mean:.3f}")
                        else:
                            print(f"      -> Skipping column (zeros  50% of total length)")

            # Update the sheet in the dictionary
            dfs[sheet_name] = df

        # Save updated data back to Excel
        with pd.ExcelWriter(output_file, engine="xlsxwriter") as writer:
            for sheet_name, df in dfs.items():
                df.to_excel(writer, sheet_name=sheet_name, index=False, float_format="%.3f")

        print(f"  -> Updated file saved: {output_file}")
```
