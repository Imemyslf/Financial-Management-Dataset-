```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\sharm\OneDrive\Desktop\Kishan\Contractzy\WebScrapping\Tutorial\Code\Rayyan\Feature Engineering\ipynb code\fe_bscfplry.xlsx"  # Update with the correct file path
xls = pd.ExcelFile(file_path)
```

```python
df = xls.parse(sheet_name="Sheet1")  

summary_stats = df.describe()
print("Summary Statistics:")
print(summary_stats)
```

```python
selected_columns = [
    "Total Shareholders Funds",
    "Total Revenue",
    "Net CashFlow From Operating Activities",
    "Enterprise Value (Cr.)",
    "Total Debt/Equity (X)"
]
```

```python
sns.set_style("whitegrid")

# Plot histograms
fig, axes = plt.subplots(3, 2, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(selected_columns):
    sns.histplot(df[col], bins=10, kde=True, ax=axes[i], color="blue")
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.savefig("eda_histograms.png", dpi=300, bbox_inches='tight')
plt.show()
```
