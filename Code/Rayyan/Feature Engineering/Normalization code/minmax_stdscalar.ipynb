```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load the standardized balance sheet dataset
df = standardized_df  # Update with your actual file path

# Compute correlation matrix
corr_matrix = df.corr()

# Visualizing the correlation matrix
plt.figure(figsize=(40, 20))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Identify highly correlated features (Threshold > 0.8)
high_corr_pairs = set()
threshold = 0.8

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            high_corr_pairs.add((corr_matrix.columns[i], corr_matrix.columns[j]))

print("\nHighly Correlated Features (> 0.8):")
for pair in high_corr_pairs:
    print(pair)


```

```python
# Drop one of each pair of highly correlated features
features_to_drop = set([pair[1] for pair in high_corr_pairs])  # Remove duplicates

df_selected = df.drop(columns=features_to_drop)

print(len(features_to_drop))
print(features_to_drop)
print(df_selected)
```

```python
# Compute VIF for remaining features
X = df_selected.copy()  # Feature dataset

print(X.head(5))
```

```python
vif_data = pd.DataFrame()

vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# print(vif_data["Feature"])
print(vif_data["VIF"])
os.makedirs(r"C:\Users\sharm\OneDrive\Desktop\Kishan\Data\VIF",exist_ok=True)
vif_data.to_excel(r"C:\Users\sharm\OneDrive\Desktop\Kishan\Data\VIF\vif_data.xlsx")

```

```python

# Drop features with VIF > 10 (high multicollinearity)
vif_threshold = 10
features_to_remove_vif = vif_data[vif_data["VIF"] > vif_threshold]["Feature"].tolist()
print(features_to_remove_vif)
```

```python
df_final = df_selected.drop(columns=features_to_remove_vif)

print("\nRemoved Features (High Multicollinearity - VIF > 10):", features_to_remove_vif)
print("\nFinal Selected Features:", df_final.columns)

# Save final feature set
df_final.to_excel("selected_features.xlsx", index=False)
print("\nFinal dataset with selected features saved as 'selected_features.xlsx'")
```
