```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume bs_df is already loaded and processed
bs_df.rename(columns={bs_df.columns[0]: 'Year'}, inplace=True)

# Set Year as index for better visualization
bs_df.set_index('Year', inplace=True)

# Transpose the dataframe to swap axes
bs_df_transposed = bs_df.T

# Plot the data
plt.figure(figsize=(12, 6))
bs_df_transposed.plot(kind='bar', figsize=(12, 6), width=0.8)

plt.xlabel('Financial Parameters')
plt.ylabel('Value (in Rs. Cr.)')
plt.title('Pruned Balance Sheet Overview')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
