```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure data is numeric
bs_numeric_df = bs_df.select_dtypes(include=[np.number])
bs_numeric_df
```

```python
# Handle NaN values (fill or drop)
bs_numeric_df = bs_numeric_df.fillna(0)

# Plot the correlation heatmap
plt.figure(figsize=(40, 20))
sns.heatmap(bs_numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Balance Sheet Data")

image_path = r"C:\Users\sharm\OneDrive\Desktop\Kishan\Contractzy\WebScrapping\Tutorial\Financial_Data\MoneyControl\Companies\IT Services & Consulting\3i Infotech Ltd\Pruned_Excel\Combined_Pruned_Excel\bs_df_corr.png"
plt.savefig(image_path, dpi=300, bbox_inches="tight")  # Save with high resolution

plt.show()
```

```python
# Get the correlation matrix
corr_matrix = bs_numeric_df.corr().abs()

# Select upper triangle of correlation matrix (to avoid duplicate pairs)
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find columns with correlation > 0.9
high_corr_features = [column for column in upper_tri.columns if any(upper_tri[column] > 0.8)]

print("Highly Correlated Features (Above 0.8):")
print(len(high_corr_features))
print(high_corr_features)
```

```python
# Drop one of each highly correlated feature pair
bs_reduced_df = bs_numeric_df.drop(columns=high_corr_features)

# print("Shape Before:", bs_numeric_df.shape)
# print("Shape After:", bs_reduced_df.shape)
bs_reduced_df.to_excel(r"C:\Users\sharm\OneDrive\Desktop\Kishan\Data\bs_more_reduced_df.xlsx",index=False)
```

```python
# Example: Checking if 'Total Assets' is the sum of 'Total Non-Current Assets' and 'Total Current Assets'
bs_numeric_df["Check_Total_Assets"] = bs_numeric_df["Total Non-Current Assets"] + bs_numeric_df["Total Current Assets"]

# Compare with original column
is_total_assets_derived = (bs_numeric_df["Check_Total_Assets"] == bs_numeric_df["Total Assets"]).all()
print("Is 'Total Assets' a derived feature?", is_total_assets_derived)

# If true, drop 'Total Assets' because it's just a sum
if is_total_assets_derived:
    bs_reduced_df = bs_reduced_df.drop(columns=["Total Assets"])
```
